```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('drug.csv', skiprows= 6)
df.drop("[run number]", axis = 1, inplace = True)
df.drop("[step]", axis = 1, inplace = True)
df.drop("surplus", axis = 1, inplace = True)
#df = df[df['no_of_consumers'] == 100]
df
```

```python
g = sns.PairGrid(df,
                 x_vars=["tolerance_limit", "no_of_dealers", "no_of_consumers", 'consumer_network', 'franchise_limit',
                        'leave_consumer', 'suspicion', 'allow_direct_connection', 'new_consumer'],
                 y_vars=["mean [tax_collected] of consumers" ,'mean [asked_someone_to_broker] of consumers', 'mean [brokered_for_someone] of consumers',
                        'mean [brokered_for_new_user] of consumers', 'mean [pinched] of consumers', 'mean [selling_price] of dealers',
                        'count dealers', 'count consumers'],
                 aspect=.75, size=4.0)
g.map(sns.violinplot, palette="pastel");
```
